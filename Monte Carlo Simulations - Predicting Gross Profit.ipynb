```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# rev_m = mean of revenues
# rev_stdv = standard deviation of revenues

rev_m = 170
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
# Cogs = 60% of revenues approximately
# may deviate 10% on the mean of CoGS (Standard Deviation)
COGS = -(rev*np.random.normal(0.6,0.1))

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show
```
